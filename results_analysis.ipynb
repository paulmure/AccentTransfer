```python
import numpy as np
from matplotlib import pyplot as plt
import pickle
```

```python
with open('training_log', 'rb') as f:
    training_log = pickle.load(f)
```

```python
training_log['epoch_1'].keys()
```

```python
epoch_recon_losses = [np.mean(logs['loss_recons']) for _, logs in training_log.items()]
plt.plot(epoch_recon_losses)
plt.title('reconstruction loss as each epoch')
plt.savefig('recon_loss_at_each_epoch')
```

```python
all_recon_losses = [recon_loss for _, logs in training_log.items() for recon_loss in logs['loss_recons']]
plt.plot(all_recon_losses)
plt.title('all reconstruction loss')
plt.savefig('all_recon_loss')
```

```python
epoch_multitask_losses = [np.mean(logs['loss_multitask']) for _, logs in training_log.items()]
plt.plot(epoch_multitask_losses)
plt.title('multitask loss as each epoch')
plt.savefig('multitask_loss_at_each_epoch')
```

```python
all_multitask_losses = [recon_loss for _, logs in training_log.items() for recon_loss in logs['loss_multitask']]
plt.plot(all_multitask_losses)
plt.title('all multitask loss')
plt.savefig('all_multitask_loss')
```

```python
epoch_total_loss = [np.mean(logs['total_loss']) for _, logs in training_log.items()]
plt.plot(epoch_total_loss)
plt.title('total_loss as each epoch')
plt.savefig('total_loss_at_each_epoch')
```

```python
all_total_loss = [recon_loss for _, logs in training_log.items() for recon_loss in logs['total_loss']]
plt.plot(all_total_loss)
plt.title('all total_loss')
plt.savefig('all_total_loss')
```
